# 필요 개념들


### Norm
* Norm 은 벡터의 크기를 측정하는 방법
* 두 벡터 사이의 거리를 측정하는 방법이기도 합니다.


![norm공식](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FbwhJNd%2FbtqT03hLtI8%2FpifbXPF3dRyxDTrVveIfK0%2Fimg.png)

 > 머신러닝에서 norm이란 유사도 계산를 위해 사용하고 데이터의 특성을 측정하고 모델의 학습 및 평가에 중요한 역할을 수행한다



### L1 Norm
* L1 Norm 은 벡터 p, q 의 각 원소들의 차이의 절대값의 합

![L1 Norm](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2Fco7vji%2FbtqTYRvlYnz%2Fsk2VP94MX9q7xzND0EA2sk%2Fimg.png)


두 벡터 p =(6, 9, -2), q = (3, 0, 7) 의 L1 Norm을 구한다면
|6-3| + |9-0| + |-2 -7| = 3 + 9 + 9 = 21


### L2 Norm
* L2 Norm 은 벡터 p, q 의 유클리디안 거리(직선 거리) 입니다.

![L2 Norm](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FbhAQWT%2FbtqTSTf9bR7%2FBCu9Cdg6tijZEGdwyEoDkK%2Fimg.png)

* 여기서 q 가 원점이라면, 벡터 p, q의 L2 Norm 은 벡터 p와 원점 간의 직선거리라고 할 수 있습니다.
* 예를 들어, 두 벡터 p =(3, 1, -3), q = (0, 0, 0) 의 L2 Norm 을 구한다면
3^2 + 1^2 + (-3)^2 = 19


> 여기서 유클리디안 이란??
### 유클리디안 거리 기반 유사도
* 두 점 사이의 거리를 측정하는 유클리디안 거리를 사용하여 문서의 유사도를 구하는 방법으로 거리가 가까울수록 유사도가 높다고 판단함
* 자카드 유사도나 코사인 유사도만큼 유용하게 사용되는 방법은 아니나, 자연어 퍼리 분야뿐만 아니라 다른 분야에서도 범용적으로 사용되는 거리 측정 기법

### L1 Norm과 L2 Norm 의 차이
* 검정색 두 점사이의 L1 Norm 은 빨간색, 파란색, 노란색 선으로 표현 될 수 있고
* L2 Norm 은 오직 초록색 선으로만 표현될 수 있습니다.
* L1 Norm 은 여러가지 path 를 가지지만 L2 Norm 은 Unique shortest path 를 가집니다.
* 예를 들어 p = (1, 0), q = (0, 0) 일 때 L1 Norm = 1, L2 Norm = 1 로 값은 같지만 여전히 Unique shortest path 라고 할 수 있습니다.

### L1 Loss

![L1 Loss](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FbGnGJG%2FbtqTPxRVxF0%2FvcfUpEEVKCJUt4UtNRbM11%2Fimg.png)

* y_i 는 실제값을, f(x_i)는 예측값을 의미합니다.
* 실제값과 예측값 사이의 오차의 절대값을 구하고, 그 오차들의 합을 L1 Loss 라고 합니다.
*  L1 Loss는 실제값과 예측값 사이의 차이를 절댓값으로 측정하기 때문에, 이상치(outlier)에 민감하게 반응하고 큰 오차에 더 많은 영향을 받는다


### L2 Loss
![L2 Loss](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FEv82h%2FbtqT0tuaiu5%2FruUT2eTiYLhRWGQEVTYeHk%2Fimg.png)
* L2 Loss 는 오차의 제곱의 합으로 정의됩니다.
*  L2 Loss는 실제값과 예측값 사이의 차이를 제곱하여 측정하기 때문에, 이상치에 덜 민감하게 반응하고 작은 오차에 대해서도 주의를 기울입니다.



 > L1 Loss는 이상치가 많이 포함된 데이터셋에서 사용될 수 있으며, 모델이 이상치에 크게 영향을 받을 수 있습니다. 반면에 L2 Loss는 이상치에 대해서 상대적으로 덜 민감하게 작용하며, 오차의 크기를 좀 더 균형있게 고려하는 경향을 가지고 있습니다. 두 손실 함수 중 어떤 것을 사용할지는 문제의 특성과 데이터의 특징에 따라 결정됩니다.



### L1 정규화, L2 정규화

* 정규화란?
  * -> 모델의 복잡성을 제어하고 일반화 성능을 향상시키기 위해 모델의 가중치를 제한하는 방법

* 사용하는 이유
 * -> 모델에서 학습을 진행하는하는데 특정 데이터에 따라 가중치가 커지게 되므로 과적합을 막기 위해 사용한다





* L1 정규화
  * L1 정규화는 모델의 가중치의 절댓값에 비례하는 패널티를 부과하여 가중치 값을 제한
  ![L1 정규화](https://oopy.lazyrockets.com/api/v2/notion/image?src=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FbkClJn%2FbtqLA20hMN9%2FKR9d27A0SkCKkad91TPujk%2Fimg.png&blockId=d4c39c4d-5e44-45c8-bf32-43d1085b2709)
  * L1 정규화는 모델의 가중치를 0으로 만들어 피처 선택(Feature Selection)의 역할을 할 수 있습니다. 이는 일부 피처들이 모델에 미치는 영향력이 작아지거나 없어지게 합니다. 따라서, L1 정규화는 희소성(Sparsity)을 갖는 모델을 만들 수 있으며, 중요한 피처들을 자동으로 선택할 수 있습니다
  
  > 결과적으로 몇몇 중요한 가중치들만 남게 된다.


* L2 정규화
  * L2 정규화는 모델의 가중치의 제곱에 비례하는 패널티를 부과하여 가중치 값을 제한합니다.
  ![L2 정규화 이미지](https://oopy.lazyrockets.com/api/v2/notion/image?src=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FnBcsT%2FbtqLBMv9S87%2FbRKUoh94BcqjwN9Nm2xSi0%2Fimg.png&blockId=3a71f977-4b9c-41f4-a9e7-0d35f344b426)
  * 손실 함수에 제곱한 가중치를 더함으로써 편미분을 통해 역전파(Backpropagation)를 수행할 때, L2 정규화(L2 Regularization)를 적용한 경우, 비용 함수(Cost function)를 편미분하는 과정에서 가중치(weight) 뿐만 아니라 비용 함수에 대한 가중치의 미분값도 감소합니다. 이를 통해 특정 가중치가 비이상적으로 커지는 상황을 방지하고 가중치 감쇠(weight decay)가 가능해집니다
  .
  * 역전파(Backpropagation)는 인공신경망(Artificial Neural Network)에서 가중치와 편향을 조정하기 위해 사용되는 학습 알고리즘
> 전체적으로 가중치를 작아지게 하여 과적합을 방지.

* L1 Loss와 L1 정규화는 Sparse 모델을 유도하고 특정 가중치를 0에 가깝게 만들 수 있으며, L2 Loss와 L2 정규화는 가중치의 크기를 작게 유지하여 과적합을 방지하고 일반화 성능을 향상시킬 수 있습니다.

### 선형 회귀 모델에의 적용
* 선형 회귀 모델에서 L1 규제를 주는 것이 lasso regression
* 선형 회귀 모델에서 L2 규제를 주는 것이 Ridge regression
* elasticnet 은 두 개 다 쓰는 것 → L1 규제와 L2 규제를 모두 쓰되 비중을 다르게 함.

L1-Norm과 L2-Norm은 벡터의 크기를 측정하는 방법으로서, L1 정규화(L1 Regularization)와 L2 정규화(L2 Regularization)와 연관성이 있습니다.

L1 정규화는 손실 함수에 L1-Norm을 가중치로 곱하여 추가하는 방식으로 작동합니다. 이는 가중치의 절댓값을 패널티로 부과하여 모델의 복잡성을 줄이고 희소한 모델을 생성하는 기법입니다. L1-Norm을 사용하는 L1 정규화는 특정 가중치들을 0으로 만들어 피처 선택(Feature Selection)과 같은 효과를 가져옵니다.

반면, L2 정규화는 손실 함수에 L2-Norm의 제곱을 가중치로 곱하여 추가하는 방식으로 작동합니다. 이는 가중치의 크기를 제한하고 모델의 복잡성을 줄이는 효과를 가지며, 가중치 감쇠(Weight Decay)라고도 불립니다. L2-Norm을 사용하는 L2 정규화는 가중치의 크기를 줄이면서도 모델의 학습이 이루어지도록 도와주는 기법입니다.

따라서, L1-Norm과 L2-Norm은 모델의 정규화에 사용되는 방법으로서, L1 정규화와 L2 정규화는 각각 L1-Norm과 L2-Norm을 이용하여 가중치에 패널티를 부과하여 모델의 복잡성을 조절합니다. 이를 통해 특정 가중치의 영향을 줄이고 모델의 일반화 성능을 향상시키는 효과를 얻을 수 있습니다. L1 정규화는 희소성과 피처 선택에 더 강조된 효과를 가지며, L2 정규화는 가중치의 크기 제한과 모델의 일반화에 더 강조된 효과를 가지게 됩니다.
